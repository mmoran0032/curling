2016 Ford World Women's Curling Championship
============================================

Using the results of the round robin play, try to predict who will win the entire bonspiel.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy
import pandas
import seaborn

seaborn.set(style='whitegrid')
```

```python
filename = 'data/2016FordWorld-women.csv'

data = pandas.read_csv(filename)
data.head()
```

```python
data.describe()
```

```python
ends = ['e{}'.format(i) for i in range(1, 12)]
scoring = ['{}scoring'.format(e) for e in ends]

def check_scores(data):
    t1scored = data[scoring] == 1
    t2scored = data[scoring] == 2
    t1scored.columns = ends
    t2scored.columns = ends
    return (((data[t1scored].sum(axis=1) != data['final1']).sum() +
            (data[t2scored].sum(axis=1) != data['final2']).sum()),
            t1scored, t2scored)

value, t1scored, t2scored = check_scores(data)
print(value)
```

```python
data['t1endsscored'] = t1scored.sum(axis=1)
data['t2endsscored'] = t2scored.sum(axis=1)

data['pointsperscoring1'] = data['final1'] / data['t1endsscored']
data['pointsperscoring2'] = data['final2'] / data['t2endsscored']

data['endsplayed'] = data[ends].notnull().sum(axis=1)
data['game_number'] = data.index
```

```python
team1won = data['final1'] > data['final2']

def determine_winning_team(x):
    return 1 if x else 2

data['winner'] = team1won.apply(determine_winning_team)
data.head()
```

OK, we should have all of the data split up into a very basic form, so we can create a dataframe where the teams in each game are split.

```python
columns = ['game', 'team_name', 'final_score', 'scoring_ends',
           'points/end', 'ends_played', 'LSFE', 'winner']

team1 = data[['game_number', 'Team1', 'final1', 't1endsscored',
              'pointsperscoring1', 'endsplayed', 'LSFE', 'winner']]
team1.columns = columns
team1.set_index(['game'], inplace=True)
team1.loc[:, 'won'] = team1.apply(lambda x: x['winner'] == 1, axis=1)
team1.loc[:, 'started_hammer'] = team1.apply(lambda x: x['LSFE'] == 1, axis=1)

team2 = data[['game_number', 'Team2', 'final2', 't2endsscored',
              'pointsperscoring2', 'endsplayed', 'LSFE', 'winner']]
team2.columns = columns
team2.set_index(['game'], inplace=True)
team2.loc[:, 'won'] = team2.apply(lambda x: x['winner'] == 2, axis=1)
team2.loc[:, 'started_hammer'] = team2.apply(lambda x: x['LSFE'] == 2, axis=1)

team1.head()
```

```python
team1[team1['won'] & team1['started_hammer']]['won'].sum() / team1['won'].count()
```

```python
teamdata = pandas.concat([team1, team2])
teamdata.drop(['LSFE', 'winner'], axis=1, inplace=True)
teamdata.shape
```

We can look at the data by team name to see if we see any trends. For our predictions, we'll also want to look at the standard deviations, since we'll need those for predicting the outcomes of the games.

```python
means = teamdata.groupby(['team_name']).mean()
stds = teamdata.groupby(['team_name']).std(ddof=0)
```
